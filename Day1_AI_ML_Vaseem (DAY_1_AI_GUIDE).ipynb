```python
# write a program to perform arithmetic operations
# (+,-,/,*) on the two number entered by user
# and the operator is also entered by user
# (operation to be performed)

# then , ask user do you want to continue?
# (y or n) if y repeat the process
# if n then exit the code with proper message
```

```python
flag=0 # initialise the flag as 0

while(flag==0):
  A = int(input("number1: "))
  B = int(input("number2: "))
  op = input("enter the operation(+,-,*,/):")

  if(op == '+'):
    C = A+B# arithmetic operators (+,-,/,*)
    print('result: ',C)
  elif(op == '-'):
    C = A-B
    print('result: ',C)
  elif(op == '*'):
    C = A*B
    print('result: ',C)
  elif(op == '/'):
    C = A/B
    print('result: ',C)
  else:
    print("Wrong operation please select (+,-,/,*)")
  chk = input("Do you want to continue (y or n)?")
  if(chk=='n' or chk == 'N'):
    flag=1# change the flag status
    print("program exited")
  elif(chk=='y' or chk == 'Y'):
    flag=0
    print("program continued")

```

```python
#functions
# input collection function

def input_op():
  A = int(input("number1: "))
  B = int(input("number2: "))
  f=0
  while(f==0):
    op = input("enter the operation(+,-,*,/):")
    if(op=='+' or op=='-' or op=='*' or op=='/'):
      f=1
      print('right operator selected')
    else:
      f=0
      print("select correct operator")
  return([A,B,op])
```

```python
# operation execution
def execution_func(input_list):
  A = input_list[0] # input 1
  B = input_list[1] # input 2
  op = input_list[2] # operation
  if(op == '+'):
    C = A+B# arithmetic operators (+,-,/,*)
    print('result: ',C)
  elif(op == '-'):
    C = A-B
    print('result: ',C)
  elif(op == '*'):
    C = A*B
    print('result: ',C)
  elif(op == '/'):
    C = A/B
    print('result: ',C)
  return([C,op])

```

```python
# continue checking

def check_execution():
  f2 = 0
  while(f2==0):
    chk = input("Do you want to continue (y or n)?")
    if(chk=='n' or chk == 'N'):
      flag=1# change the flag status
      f2=1
      print("program exited")
    elif(chk=='y' or chk == 'Y'):
      flag=0
      f2=1
      print("program continued")
    else:
      print("Wrong char selected")
      f2=0

  return(flag)

```

```python
# main program
flag=0
while(flag==0):
  input_Lst = input_op() # call the input function to
  # take inputs and operations

  execution_func(input_Lst) # call the execution function to
  # perform operation

  flag=check_execution()# check for the continuation

```

```python
import matplotlib.pyplot as plt
import numpy as np
# sin function
# cos function
# custom function
X = np.linspace(1,20,100)
Y = np.sin(X)
Z = np.cos(X)
W = 2*np.sin(X)+np.cos(X)
P = np.sin(2*X) - 2*np.cos(X)
```

```python
# all functions output in different figure
plt.figure(1,figsize=(5,2))#activting the figure number 1
plt.plot(X,Y,'r',linewidth=4) # 'r','g,'m','c','k','y','b'
plt.title("sine function")
plt.xlabel('Range')
plt.ylabel('sin(X)')
plt.grid('on')

plt.figure(2,figsize=(5,2))#activting the figure number 1
plt.plot(X,Z,'g',linewidth=4) # 'r','g,'m','c','k','y','b'
plt.title("cos function")
plt.xlabel('Range')
plt.ylabel('cos(X)')
plt.grid('on')

plt.figure(3,figsize=(5,2))#activting the figure number 1
plt.plot(X,W,'b',linewidth=4) # 'r','g,'m','c','k','y','b'
plt.title("custom function")
plt.xlabel('Range')
plt.ylabel('func(X)')
plt.grid('on')

plt.figure(4,figsize=(5,2))#activting the figure number 1
plt.plot(X,P,'k',linewidth=4) # 'r','g,'m','c','k','y','b'
plt.title("custom function")
plt.xlabel('Range')
plt.ylabel('func(X)')
plt.grid('on')
```

```python
# comparative analysis
plt.figure(1,figsize=(10,4))#activting the figure number 1
plt.plot(X,Y,'r',linewidth=4,label="sin(X)") # 'r','g,'m','c','k','y','b'
plt.plot(X,Z,'g',linewidth=4,label='Cos(X)')
plt.plot(X,W,'b',linewidth=4,label='func1(X)')
plt.plot(X,P,'m',linewidth=4,label='func2(X)')
plt.title("comparative analysis")
plt.xlabel('Range')
plt.ylabel('function(X)')
plt.grid('on')
plt.legend() # activate the legends
```
